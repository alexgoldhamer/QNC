```python
#Exercise 1

import numpy as np
from scipy.stats import binom

n=10
probability_release=0.2
k=np.arange(n+1)
probability=binom.pmf(k,n,probability_release)

for i in range(n + 1):
    print(f"P({i} quanta released) = {probability[i]:.5f}")

#Exercise 2
n=14
k=8
ps=np.arange(0,1.0,0.1)
probability=[binom.pmf(k,n,p) for p in ps]
for i in range(len(ps)):
    print(f"P({k} proability| p={ps[i]:.1f}) = {probability[i]:.5f}")

#Exercise 3
n=14
k1=8
k2=5
probability_release=0.1
probability1=binom.pmf(k1,n,probability_release)
probability2=binom.pmf(k2,n,probability_release)
total_probability=probability1*probability2
total_log_probability=np.log(probability1)+np.log(probability2)
print(f'total_probability = {total_probability:.5f}, total_log_probability = {total_log_probability:.5f}')

ps=np.arange(0,1.0,0.1)
probability1=[binom.pmf(k1,n,p) for p in ps]
probability2=[binom.pmf(k2,n,p) for p in ps]
total_probability=[probability1[i]*probability2[i] for i in range(len(ps))]
total_log_probability=[np.log(probability1[i])+np.log(probability2[i]) for i in range(len(ps))]
for i in range(len(ps)):
    print(f'P({k1} quanta| p={ps[i]:.1f}) * P({k2} quanta| p={ps[i]:.1f}) = {total_probability[i]:.5f}, log = {total_log_probability[i]:.5f}')

#Exercise 4
import numpy as np
from scipy.stats import binom

# Observed data
n = 14  # number of trials per experiment
observed_counts = np.array([0, 0, 3, 7, 10, 19, 26, 16, 16, 5, 5, 0, 0, 0, 0])  # length 15 for k=0 to 14
k_values = np.arange(len(observed_counts))

# Candidate p values from 0.00 to 1.00 (step 0.01)
p_values = np.arange(0.00, 1.01, 0.01)
log_likelihoods = []

for p in p_values:
    # Compute PMF for each k at this p
    pmf = binom.pmf(k_values, n, p)
    
    # Avoid log(0) by replacing 0 with a small number
    pmf[pmf == 0] = 1e-10

    # Log-likelihood: sum(count_k * log(P(k)))
    log_likelihood = np.sum(observed_counts * np.log(pmf))
    log_likelihoods.append(log_likelihood)

# Find p that gives the maximum log-likelihood
max_index = np.argmax(log_likelihoods)
p_hat = p_values[max_index]

print(f"Maximum Likelihood Estimate (p-hat) = {p_hat:.2f}")

#Exercise 5
import numpy as np
from scipy.stats import binomtest

# Parameters
n = 14               # number of trials (quanta)
k = 7                # observed releases
p_null = 0.3         # null hypothesis: true release probability

# Perform two-tailed binomial test
result = binomtest(k, n, p=p_null, alternative='two-sided')

# Print result
print(f"Observed: {k} out of {n} releases")
print(f"Null hypothesis p = {p_null}")
print(f"p-value = {result.pvalue:.5f}")

# Conclusion
alpha = 0.05  # significance level
if result.pvalue < alpha:
    print("Result is statistically significant — temperature likely had an effect.")
else:
    print("Result is not statistically significant — no evidence that temperature changed release probability.")
```
